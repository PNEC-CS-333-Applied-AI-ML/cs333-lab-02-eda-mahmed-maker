```python
                                                      # PART II
# CALCULATING RANGE
range_df1 = df1.select_dtypes(include='number').agg(['min', 'max'])
range_df1.loc['Range'] = range_df1.loc['max'] - range_df1.loc['min']

print(range_df1)
```

```python
                                                        # PART III
# CALCULATING IQR
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)

IQR = Q3 - Q1
print(IQR)

df1 = df1[~((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]

df1.shape
```

```python
                                                              # PART VIII
# MODIFIED Z-SCORE METHOD
from scipy.stats import median_abs_deviation

print("MODIFIED Z-SCORE OUTLIERS:")
for col in df.select_dtypes(include=[np.number]).columns:
    median = df[col].median()
    mad = median_abs_deviation(df[col])
    z_scores = 0.6745 * (df[col] - median) / mad
    outliers = df[np.abs(z_scores) > 3.5]
    print(f"{col}: {len(outliers)} outliers")

```


### âœï¸Which method was selected?
Modified Z-score method was selected.

### âœï¸Why it was selected?
Because Modified Z-score uses median and MAD (Median Absolute Deviation) which is more robust than mean and standard deviation. It works better when data is skewed or has outliers, as median is not affected by extreme values.

### âœï¸Which parameter contain significant outliers?
Tool Wear (VB) contained significant outliers. The outliers detected were: 1.07, 1.30, 1.53, and 1.14.

### âœï¸How removal affected dataset statistics?
The removal made the data more reliable for analysis by reducing the influence of extreme values on mean and standard deviation
